```python
from collections import Counter

d8 = []
for roll in range (1, 9):
    d8.append(roll)
print(d8)

d4 = []
for roll in range (1, 5):
    a = roll
    for roll2 in range (1, 5):
        b = roll2
        newpair = (a, b)
        d4.append(newpair)
print(d4)

roll_counts = Counter(d8)
print('roll counts is: ', roll_counts)
expected_damage = 0
for count in d8:
    ##Find roll value
    print("d8 value: " + str(count))
    
    ##Find probability of rolling this value
    prob_roll = (roll_counts[count]/len(d8))
    print("d8 probability: " + str(prob_roll))
    
    ##Find probability of rolling this value * value
    prob_value = prob_roll*count
    print("d8 product: " + str(prob_value))
    
    ##Sum up the products for each value
    expected_damage = expected_damage + prob_value

##Print the final product
print("d8 exp damage: " + str(expected_damage))
     
unique_combo = []
for tuple in d4:
    ##Find sum of values rolled by each of the two dice
    sum_tuple = 0
    for value in tuple:
        sum_tuple = sum_tuple + value
    unique_combo.append(sum_tuple)

expected_damage2 = 0
roll_counts2 = Counter(unique_combo)
print (roll_counts2)
for count in roll_counts2:
    ##Find the total of each roll
    print ("d4 value:" + str(count))
    
    ##Find the probability of rolling such a total
    prob_roll2 = roll_counts2[count]/len(d4)
    print ("d4 probability:" + str(prob_roll2))
    
    ##Find probability of rolling this value * value
    prob_value2 = prob_roll2*count
    print("d4 product: " + str(prob_value2))
    
    ##Sum up the products for each value
    expected_damage2 = expected_damage2 + prob_value2
    
##Print the final product
print("d4 exp damage: " + str(expected_damage2))

```

```python

##for the d8 roll
sum_all = 0
for value in d8:
    sum_all = sum_all + value
    
x_avg = sum_all/len(d8)

d8_squares = [(x-x_avg)**2 for x in d8]
print(d8_squares)

sum_squares = 0
for square in d8_squares:
    sum_squares = sum_squares + square

d8_variance = sum_squares/len(d8_squares)

print (d8_variance)

##for the d4 roll
d4_sum_all = 0
total_values = []
for tuple in d4:
    sum_tuple = 0
    for d4_value in tuple:
        sum_tuple = sum_tuple + d4_value
    total_values.append(sum_tuple) ##make a list of all possible values by the 2 dice
    d4_sum_all = d4_sum_all + sum_tuple

print(total_values)

d4_x_avg = d4_sum_all/len(d4)

d4_squares = []
for value in total_values:
    s = (value-d4_x_avg)**2
    d4_squares.append(s)

d4_sum_squares=0
for square in d4_squares:
    d4_sum_squares = d4_sum_squares + square

print(d4_squares)

d4_variance = d4_sum_squares/len(d4_squares)
print(d4_variance)
```
